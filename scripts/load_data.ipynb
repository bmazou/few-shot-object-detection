```python
import pandas as pd
import os
```

```python
classes = ['Apple', 'Car', 'Dog', 'Cat']
```

```python
training_annotation_loc = 'csv_folder/validation-annotations-bbox.csv'
training_annotation_file = pd.read_csv(training_annotation_loc)
```

```python
# The class-descriptions-boxable.csv contains the name of all 600 classes with their corresponding ‘LabelName’
class_descriptions_file='csv_folder/class-descriptions-boxable.csv'
class_descriptions = pd.read_csv(class_descriptions_file, header=None)
class_descriptions.head()
```

```python
# Find the label_name for 'Apple', 'Orange' and 'Light Switches' classes
apple_pd = class_descriptions[class_descriptions[1]=='Apple']
car_pd = class_descriptions[class_descriptions[1]=='Car']
dog_pd = class_descriptions[class_descriptions[1]=='Dog']
cat_pd = class_descriptions[class_descriptions[1]=='Cat']

apple_pd


label_name_apple = apple_pd[0].values[0]
label_name_car = car_pd[0].values[0]
label_name_dog = dog_pd[0].values[0]
label_name_cat = cat_pd[0].values[0]

label_name_cat
```

```python
apple_bbox = training_annotation_file[training_annotation_file['LabelName']==label_name_apple]
car_bbox = training_annotation_file[training_annotation_file['LabelName']==label_name_car]
dog_bbox = training_annotation_file[training_annotation_file['LabelName']==label_name_dog]
cat_bbox = training_annotation_file[training_annotation_file['LabelName']==label_name_cat]
```

```python
apple_img_id = apple_bbox['ImageID']
car_img_id = car_bbox['ImageID']
dog_img_id = dog_bbox['ImageID']
cat_img_id = cat_bbox['ImageID']
```

```python
df = pd.concat([apple_bbox, car_bbox, dog_bbox, cat_bbox], axis=0)
df
```

```python
def get_img_names(classes):
    img_names = []
    for thing in classes:
        path = 'Dataset/validation/' + thing + '/Label'
        files = os.listdir(path)
        
        for i in range(len(files)):
            files[i] = files[i].split('.')[0]
            
        img_names += files
        
    return img_names

imgs = get_img_names(classes)
len(imgs)
```

```python
imgs
```

```python
# Filter df to only include desired images
new_df = df.drop(df[ ~df['ImageID'].isin(imgs) ].index)
new_df
```
